# Import Libraries


```python
import pandas as pd
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

# Create Dataset to Practice






```python

# Generate synthetic data
np.random.seed(42)

data = {
    "ProductID": [f"P{i}" for i in range(1, 21)],
    "Category": np.random.choice(["Electronics", "Clothing", "Home", "Toys"], size=20),
    "Price": np.random.randint(10, 500, size=20),
    "UnitsSold": np.random.randint(50, 2000, size=20),
    "Rating": np.random.uniform(2.5, 5.0, size=20),
    "Feature1": np.random.randint(1, 100, size=20),
    "Feature2": np.random.randint(1, 100, size=20),
    "Feature3": np.random.randint(1, 100, size=20),
}

df = pd.DataFrame(data)
df
```

# Select relevant features for clustering


```python
features = df[["Price", "UnitsSold", "Rating", "Feature1", "Feature2", "Feature3"]]
```

# Standardize the data


```python
# Standardize the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
Z = linkage(scaled_features, method="ward")
```

```python
plt.figure(figsize=(10, 7))
dendrogram(Z, labels=df["ProductID"].values, leaf_rotation=90)
plt.title("Dendrogram for Product Clustering")
plt.xlabel("Products")
plt.ylabel("Distance")
plt.show()
```

# Interpretation:

## P7, P1, P11, P8, P15 are grouped together at a low distace suggesting high similarity.
